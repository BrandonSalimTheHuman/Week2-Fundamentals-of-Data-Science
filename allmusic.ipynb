```python
# import stuff
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# Downloading website data
# Use user agent to be able to access data
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
# url of website
url = 'https://www.allmusic.com/'
# get html data using beautifulsoup
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# create empty lists for the columns
artists = []
titles = []
links = []
genres = []
reviews = []

# target every individual new release song
songs = soup.findAll('div', class_='new-release')

# for each song
for song in songs:
  # get base text for each column, then format it
  artist = song.find('div', class_='artist').text.replace("\n", "")
  title = song.find('div', class_='title').text.replace("\n", "")
  genre = song.find('div', class_='genres').text.replace("\n", "")
  review = song.find('div', class_='headline-review').text.replace("\n", "").strip().replace("                                                                                              ", "")
  link_to_song = song.find('div', class_='title').find('a').get('href').replace("\n", "")

  # append it to their respective lists
  artists.append(artist)
  titles.append(title)
  genres.append(genre)
  reviews.append(review)
  links.append(link_to_song)

# create dataframe using the columns and their respective lists
dataframe = pd.DataFrame(
    {
        'Title': titles,
        'Artist': artists,
        'Genre': genres,
        'Review': reviews,
        'Song link': links
    }
)

# convert to csv
dataframe.to_csv('allmusic.csv', index=False)

# show dataframe
print(dataframe)
```
